```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/bigDataSem5/employee_performance.csv')
df.head()
```

```python
top10 = df.sort_values(by='KPI_Score', ascending=False).head(10)
print("\nTop 10 Employees by KPI Score:")
print(top10[['Employee_ID', 'Name', 'Department', 'KPI_Score']])
```

```python
dept_avg = df.groupby('Department')['KPI_Score'].mean().reset_index()
print("\nAverage KPI Score by Department:")
print(dept_avg)
```

```python
df['KPI_per_Hour'] = df['KPI_Score'] / df['Hours_Worked']
print("\nPreview with KPI_per_Hour:")
print(df[['Employee_ID', 'Name', 'KPI_Score', 'Hours_Worked', 'KPI_per_Hour']].head())
```

```python
output_path = '/content/drive/MyDrive/bigDataSem5/employee_performance_cleaned.csv'
df.to_csv(output_path, index=False)
```
